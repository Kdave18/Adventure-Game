```python
print("Welcome to your own Adventure game")
# This code prompts the user to enter their name and then welcomes them to the adventure
name = input("Type your name: ")
print("Welcome",name,"to this adventure")
# This code takes the user's input to navigate through an adventure game with choices.

# Prompt the user to choose which direction to go on a dirt road
answer = input("You’re on a dirt road that has come to an end, and you can either go left or right. Which way would you choose? " ).lower()
# If the user chooses left, the game continues with the next decision
if answer == "left":
 # Prompt the user to decide whether to walk around or swim across a river
    answer = input("You come to a river. You can either walk around it or swim across. Type 'walk' to walk around or 'swim' to swim across. ")
    if answer == "swim":
      print("You swam across and were eaten by an alligator.")
    elif answer == "walk":
      print("You walked for many miles, ran out of water, and lost the game.")
# If the user enters an invalid option, they lose
    else:
      print("Not a valid option, you lose ")

# This part of the code handles the right turn when the user chooses it.
elif answer == "right":
# Prompt the user to decide whether to cross a wobbly bridge or head back
  answer = input("You come to a bridge that looks wobbly. Do you want to cross it or head back? (Type 'cross' to cross or 'head back' to go back.) ")
# If the user chooses to head back, they lose the game
  if answer == "head back":
      print("You go back and lose the game.")
# If the user chooses to cross the bridge, the game continues with the next choice
  elif answer == "cross":
# Prompt the user to decide whether to talk to the stranger or not
       answer = input("You cross the bridge and meet a stranger. Do you want to talk to them? (Yes/No) ").lower()
# If the user chooses to talk to the stranger, they win the game
       if answer == "yes":
        print(" You talk to a stranger, and they give you a food. You win the game!")
 # If the user chooses not to talk to the stranger, they lose
       elif answer == "no":
        print("You ignore the stranger and miss the food. You lose")
 # If the user enters an invalid option, they lose
       else:
        print("Not a valid option, you lose ")
 # If the user enters an invalid option when deciding to cross or head back, they lose
else :
   print("Not a valid option, you lose ")

print("Thank you for playing", name)
```
